```python
# data generation and fitting using smoothing spline plus how to plot them
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from csaps import csaps
from numpy import zeros

#data generation: note how we introduce noise?
np.random.seed(1234)

n=30
x = np.linspace(-5., 5., n)

# define the underlying function of the data
f = lambda x:x*np.exp(-(x/2.5)**2) 

# adding noise
mean = 0.2
var = 0.8
y = f(x) + (np.random.rand(n) - mean) * var 

# specify where we need to plot the model
xs = np.linspace(x[0], x[-1], 150) 

# methods with different degree of freedom controled by "smooth" parameter
ys1 = csaps(x, y, xs, smooth=0)
ys2 = csaps(x, y, xs, smooth=0.1)
ys3 = csaps(x, y, xs, smooth=1)
ys4, smooth = csaps(x, y, xs) # without smooth parameter as input, an optimized parameter is calculated and used
print('optimized smooth parameter:', smooth)

figure(figsize=(6, 4), dpi=160)
plt.plot(x, y, 'orange', marker="o", markersize=8, linestyle = 'None')
plt.plot(xs, ys1, '-r')
plt.plot(xs, ys2, '-.g')
plt.plot(xs, ys3, '--k')
plt.plot(xs, ys4, '-y')
plt.show()
```

```python
#bootstrap re-sampling: input are X,y and output are reduced sample of them (input sample size - size_rect)
def _draw_bootstrap_sample(rng, X, y):
    size_rdct = 10
    sample_indices = np.arange(len(y))
    bootstrap_indices = rng.choice(sample_indices, size=len(y)-size_rdct, replace=False)
    return X[bootstrap_indices], y[bootstrap_indices]
```

```python
#Function returning the effective degree of freedom for a training data (predictor X only) and a smooth parameter p
def effect_df(X,p):
    n = len(X)
    Y = np.identity(n)
    R = zeros([n,n])
    for i in range(n):
        yb=Y[:,i];
        R[:,i]=csaps(X,yb,X,smooth=p)
    return R.trace()
```

### The Computation of test MSE, Squared-Bias, and Variance
In order to calculate Bias$^2$($\hat{f})$, Var($\hat{f})$, we need to obtained $m$ training data sets by resampling. \
For a given data $(x,y)$ in the test set, let $\bar{\hat{y}}$ be the average of predicted $y$ at $x$ using models  obtained from different training dataset\
MSE($x$) = $\frac{1}{m}\sum\limits_{i=1}^m(y-\hat{y}_i)$ \
Bias$^2$($\hat{f}(x))= (\bar{\hat{y}}-y)^2$ \
Var$(\hat{f}(x)) = \frac{1}{m}\sum\limits_{i=1}^m(\bar{\hat{y}}-\hat{y}_i)$ \
For the total test dataset, average of values at individual $x$ will be used.
