### Tutorial of Smoothing Spline using csaps can be found at 
https://csaps.readthedocs.io/en/latest/tutorial.html

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from csaps import csaps
from numpy import zeros

np.random.seed(1234)

n=30
x = np.linspace(-5., 5., n)
y0 = x*np.exp(-(x/2.5)**2)
y = y0 + (np.random.rand(n) - 0.2) * 0.8
xs = np.linspace(x[0], x[-1], 150)

ys1 = csaps(x, y, xs, smooth=0)
ys2 = csaps(x, y, xs, smooth=0.1)
ys3 = csaps(x, y, xs, smooth=1)
ys4, smooth = csaps(x, y, xs)
print('optimized smooth parameter:', smooth)

figure(figsize=(8, 6), dpi=160)
plt.plot(x, y, 'orange', marker="o", markersize=10, linestyle = 'None')
plt.plot(x, y0, '-', 'brown')
plt.plot(xs, ys1, '-r')
plt.plot(xs, ys2, '-.g')
plt.plot(xs, ys3, '--k')
plt.plot(xs, ys4, '-b')
plt.show()

# Find the effective degree of freedom
Y = np.identity(n)
R = zeros([n,n])
for i in range(n):
    yb=Y[:,i];
    R[:,i]=csaps(x,yb,x,smooth=0)
print('effective degree of freedom:',R.trace())
```
